### **Safety Engineering**
Safety engineering is the application of basic scientic and technical principles to mitigation or reduction of loss of life, property, and the environment by recognized hazards (potential source of harm). It requires a broad knowledge of mathematics, chemistry, physics, and other basic sciences and a familiarity with one or more of the recognized branches of engineering.

### **System safety**
System safety is an engineering discipline, according to MIL_STD_882E defines system safety as "the application of engineering and management principles, criteria, and techniques to achieve acceptable risk, within the constraints of operational effectiveness and suitability, time and cost, throughout all phases of the system life cycle.

## **Standards, the machinery regulation and CE marking**
- Danish standards


Standards make the world operate more smoothly (things like staircase step size, card sizes and parking distribution). The standards can be seen everywhere in Denmark, the window standard, the bottle standard, how the flag should be hanging, chairs and desks at schools.

Standards are very important, since they provide the essential requirements, specifications, guidelines and characteristics that can be used consistently to ensure that meterials, products, processes and services are fit for their purpose. You can think of it as a common "language" that makes sure everyone understands something equily.

Its very common to use standards, to ensure that products comply with regulations, but it isn't required by law. But if you choose to not use standards, you will have a difficult time providing the specifikations needed to not use them, since they still need to live up to requirements.

Standards are a voluntary market instrument, consensus-dirven (agreed upon by a group in a comitee) and adopted by a recognized body. Standards can set requirements, example; construction (paper formats, threads, data formats), systems (quality management, risk assessment, environmental management), performance (breaking strength, safety, ergonomics, noise, radiation) and so on.

A standard itself is standardized (what?), meaning each standard has an introduction, preface, a scope of which it is applied for, normative references, definitions, symbols and abbreviations, requirements (functional requirements, e.g. test methods), annex A (normative), annex B (informative).

Two types of standards exist, De Jure and de facto standards. De jue are formal standards, voluntary, consensus-based, created in an open process and approved by a formal standardization orginzation. De facto is consortia / market dominance, industry standards, non-conensus-based and follows a closed process.

Standards are drawn up in committees and working groups consisting of experts with an interest in the subject of the standard. A committee has an average of ten members from different industries in DK, and works with a portfolio of standards appropriate to the technical area of the committee. Some standards are under revision, others are entirely new standards. Standards have to be revised every 5 years, due to technology advancing so fast.



## **CE marking and the Machinery Regulations**
The CE marking is a certification mark that indicates conformity with health, safety, and environmental protection standards for products sold within the European Economic Area (EEA*). The CE marking is also found on products sold outside the EEA that are manufactured in, or designed to be sold in, the EEA. This makes the CE marking recognizable worldwide even to people who are not familiar with EEA.
- The CE mark doesnt mean quality, but only that it meets the essential requirements, regarding safety, health and environment.

## **Accident-Causation Theories**

The most widely known theories of accident causation is:
- Domino Theory
    - Domino THeory says that 88 pct. of accidents are caused by unsafe acts committed by fellow workers, 10 pct. of accidents are caused by unsafe conditions and 2 pct. of accidents are unavoidable. It laid foundation for Axioms of Industrial Safety. The Axioms of Industrial Safety is essentially 10 main points describing why accidents occur and how they occur, as well as who are the key-players in preventing accidents.
    - Essentially there are five factos in sequence leading to an accident:
        1. **Ancestry and social environment.** - Warehouse.
        2. **Fault of a person.** - The worker.
        3. **Unsafe act/mechanical or physical hazards.** - Lifting heavy loads by hand.
        4. **Accident.** - Wrongfully lifting, using their back instead of legs.
        5. **Injury or damage** - Strained muscles, back injury.
    
    - Essentially providing the foundation for accident prevention mesaures aimed at preventing unsafe acts or unsafe conditions.
- Human Factors Theory
    - Human Factors Theory attributes accidents to a chain of events ultimately caused by human error. It consits of three broad factors that lead to human error:
        1. Overload.
        2. Inappropriate Response or Incompatibility.
        3. Inappropriate Activities.
    - **Overload**: An imbalance between the capacity of an individual at any time and the load he/she is carrying in a given state.
    - **Inappropriate Response or Incompatibility**: How a person respons to a given situation can cause or prevent an accident: e.g.:
        - A person complety disregards the specified safety procedure.
        - A person removes a safeguard from a machine for increasing output.
        - A person detects a hazardous condition, but takes no corrective action.
    - **Inappropriate Activies**: Inappropriate activites undertaken due to human error, poor judgement about the degree of risk and judgement about the degree of risk involved in a given taskand subsequenctly acting on that misjudgement.
        - Example: A person who undetakes a task that he/she does not know how to do.
